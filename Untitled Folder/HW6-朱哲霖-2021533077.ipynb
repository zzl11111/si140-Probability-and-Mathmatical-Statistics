```python
x=[0 for i in range(5000)]
s=0
var=0
countindex=[0 for i in range(4)]
for i in range(5000):
    sum=1
    count=0
    u=random.uniform(0,1)
    while(sum*u>=math.exp(-1)):
        sum=sum*u
        count+=1
        u=random.uniform(0,1)
    x[i]=count
    if(count<=3):
        countindex[count]+=1
    var+=(count-1)**2
    s+=count
plt.xlabel("3")
bins=[i for i in range(8)]
n,bins,patches=plt.hist(x,bins=bins,density=True)
mean=s/5000
var/=5000
plt.plot(bins[1:],n)
print("mean={},var={}".format(mean,var))
print("P(N=0)={} P(N=1)={} P(N=2)={},P(N=3)={}".format(countindex[0]/5000,countindex[1]/5000,countindex[2]/5000,countindex[3]/5000))
plt.savefig("./3.jpg")
```

```python
def getp(n):
    grid=np.zeros((n,n),dtype=int)

    winflag=0
    def checkwin():
        for i in range(n):
            if(grid[n-1][i]==1):
                visitlist=[0 for i in range(n*n)]
                if(checkpath(n-1,i,visitlist)):
                    return 1
        return 0
    def checkpath(row,col,visitlist):
        global winflag
        visitnumber=n*row+col
        if(visitlist[visitnumber] or grid[row,col]==0):
            visitlist[visitnumber]=1
            return 0
        if(row-1==0 and grid[row-1][col]==1):
            visitlist[visitnumber]=1
            winflag=1
            return 1
        visitlist[visitnumber]=1
        if(row+1<n):
            if(checkpath(row+1,col,visitlist)):
                return 1
        if(row-1>=0):
            if(checkpath(row-1,col,visitlist)):
                return 1
        if(col+1<n):
            if(checkpath(row,col+1,visitlist)):
                return 1    
        if(col-1>=0):
            if(checkpath(row,col-1,visitlist)):
                return 1  



    winflag=0
    count=0
    sample_count=30
    for i in range(0,sample_count):
        winflag=0
        grid=np.zeros((n,n),dtype=int)
        remain=[i for i in range(n*n)]
        while(not checkwin()):
            s=random.choice(remain)
            remain.remove(s)
            count+=1
            r=int(s/n)
            c=s%n
            grid[r][c]=1
    print(count/(sample_count*n*n))
    
    
```
