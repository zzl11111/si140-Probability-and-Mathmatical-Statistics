```python
def Alias(area_ratio):
    N=len(area_ratio)
    small=[]
    larger=[]
    accept=[0 for i in range(N)]
    alias=[0 for i in range(N)]
    area_ratio_s=np.array(area_ratio)*N

    for i,j in enumerate(area_ratio_s):
        print(i,j)
        if j<1.0:
            small.append(i)
        else:
            larger.append(i)## to divide two sets into small and large
    while larger and small:
        small_idx=small.pop()
        larger_idx=larger.pop()
        accept[small_idx]=area_ratio_s[small_idx]
        accept[larger_idx]=area_ratio_s[larger_idx]-1+area_ratio_s[small_idx]
        alias[small_idx]=larger_idx
        if(accept[larger_idx]<1.0):
            small.append(larger_idx)
        else:
            larger.append(larger_idx)
    while larger:
        larger_idx=larger.pop()
        accept[larger_idx]=1
    while small:
        small_idx=small.pop()
        accept[small_idx]=1
    return accept,alias    
```

```python
x=[0 for i in range(1000000)]
for i in range (1000000):
    count=0
    N=np.random.uniform(0,1)
    for j in range(0,20):
        if N>=0.5:
            count+=1
        N=np.random.uniform(0,1)
    x[i]=count
plt.xlabel("2-b")
n,bins,patches=plt.hist(x,bins=30,density=True)
```

```python
x=[0 for i in range(1000000)]
for i in range (1000000):
    count=0
    N=np.random.uniform(0,1)
    while N<=0.5:
        count+=1
        N=np.random.uniform(0,1)
    x[i]=count
plt.xlabel("2-c")
n,bins,patches=plt.hist(x,bins=50,density=True)
```

```python
x=[0 for i in range(1000000)]
for i in range (1000000):
    success_count=0
    failure_count=0
    N=np.random.uniform(0,1)
    while success_count<10:
        if(N<0.5):
            failure_count+=1
        else:
            success_count+=1
        N=np.random.uniform(0,1)
    x[i]=failure_count
plt.xlabel("2-d")
n,bins,patches=plt.hist(x,bins=100,density=True)
```

```python
x=[0 for i in range(5000)]
s=0
var=0
countindex=[0 for i in range(3)]
for i in range(5000):
    sum=1
    count=0
    u=np.random.uniform(0,1)
    while(sum*u>=math.exp(-1)):
        sum=sum*u
        count+=1
        u=np.random.uniform(0,1)
    x[i]=count
    if(count<3):
        countindex[count]+=1
    var+=(count-1.0032)**2
    s+=count
plt.xlabel("3")
n,bins,patches=plt.hist(x,bins=[0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5],density=True)
mean=s/5000
var/=5000

print("mean={},var={}".format(mean,var))
print("P(N=0)={} P(N=1)={} P(N=2)={}".format(countindex[0],countindex[1],countindex[2]))
```
